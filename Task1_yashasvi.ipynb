## PART I. TO IDENTIFY RECORDS PRESENT IN DATA1 BUT MSSING IN DATA2

```python

primaryKey = ["Order ID", "Product ID"]

#PERFORMING outer join of the two datafframes
#indicator makes a column "_merge" WHICH SHOWS presence of elements
df_merged = pd.merge(df1, df2, on = primaryKey, how = "outer", indicator = True)
```

```python
df_merged.tail(20)
```

```python
missingRecords1 = df_merged[df_merged["_merge"] == "left_only"]
missingRecords1.head(10)
```

```python
#ans1
len(missingRecords1)
```

### there are 507 records missing in data 2 which are present in data1


```python
print(len(df1))
print(len(df2))
```

## PART II. TO IDENTIFY RECORDS PRESENT IN DATA2 BOT MSSING IN DATA1

```python
#2nd part
missingRecords2 = df_merged[df_merged["_merge"] == "right_only"]
missingRecords2.head(10)
```

```python
#ans2.
len(missingRecords2)
```

### there are 508 records missing in data1 which are present in data2
